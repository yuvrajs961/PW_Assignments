```python
# Question 1: Discuss string slicing and provide examples?
# Answer 1: String slicing: It allows you to extract a substring from a string by specifying a range of indices in Python.
# It's done using the syntax as follow:
# string [start: stop: step]
# Some of the examples are as below:
# Basic Slicing:

string1 = "I am a good student"
string1 [0:4]

# Start and Stop:

string1 [0:]

# Negative Indexing:

strings1 [: -3]

strings2 [-3:]

# Using the step Parameter:

string1 [: :2]

# Reversing a String:

string1 [:: -1]

# Slicing with Step:
	String1 [0:5:1]

# Question 2: Explain the key features of lists on Python?
# Answer 2: A list is a versatile and a powerful data structure that can store multiples items within Python. Some of the key features are as follows:
# Ordered: It maintain the order of items, which means the elements have a defined sequence.
# Syntax:
my_list = [1, 2, 3]
# Mutable: A List can be modified after creation. We can change, add, or remove items.
# Syntax:
my_list [1] = 10
# Heterogeneous Elements: A list contains items of different types, such as integers, strings, floats, etc.
# Syntax:
my_list = [1, "apple", 3.5]
# Dynamic Size: We can grow or shrink the list in size as you add or remove elements.
# Syntax:
my_list.append (4)
# Indexed: We can also access elements using their index. Indexing starts from 0.
#Syntax:
print(my_list[0])
# Slicing: If we want a portion of the list, we can also retrieve it using slicing.
#Syntax:
print(my_list[1:3])
#Nested Lists: Lists can contain other lists which are known as nested lists.
#Syntax:
my_list = [[1, 2], [3, 4]]


# Question 3: Describe how to access, modify, and delete elements in a list with examples?
# Answer 3: We can access, modify, and delete elements in a Python list using the following methods:
# Accessing Elements
# We can access list elements using their index where Indexing starts at 0.
# Single Element Access:
# Syntax
my_list = [10, 20, 30, 40]
print(my_list[1])
# Accessing with Negative Indexing: Negative indices access elements from the end of the list.
# Syntax
print(my_list[-1])
# Slicing: We can access a range of elements through slicing.
# Syntax
print (my_list [1:3])
# Modifying Elements
# We can change the value of a specific list element by assigning a new value to an index.
# Modify a Single Element:
# Syntax
my_list[2] = 50
print(my_list)
# Modify Multiple Elements:
# Syntax
my_list [1:3] = [60, 70]
print(my_list)
# Deleting Elements
We can delete elements from a list using various methods.
# Using del Keyword:
# Syntax
del my_list [1]
print(my_list)
# Using remove () Method (removes the first matching value):
# Syntax
my_list.remove (70)
print(my_list)
# Using pop () Method (removes and returns element at a specified index and if no index is given, it removes the last element):
# Syntax
popped_value = my_list.pop()
print(my_list) # Output: [10]
print (popped_value)
# Example:
# Syntax
my_list = [1, 2, 3, 4]
print (my_list [0]) # Accessing: Output 1
my_list [1] = 20 # Modifying second element
del my_list [2] # Deleting the third element

# Question 4: Compare and contrast tuples and lists with examples?
# Answer 4: Tuples and lists: They both are sequence data types in Python, but they have key differences. Some of the differences are as follow
# Mutability
# List: Mutable — We can modify, add, or delete elements after creation.
# Tuple: Immutable — Once we create tuple you cannot modify, add, or delete elements.
# Example:
my_list = [1, 2, 3]
my_list[0] = 10  # Modifies the first element
print(my_list)

my_tuple = (1, 2, 3)
my_tuple[0] = 10  # Error! Tuples are immutable
# Syntax
# List: A list is defined using square brackets [ ].
# Tuple: A tuple is defined using parentheses ().
# Example:
my_list = [1, 2, 3]
my_tuple = (1, 2, 3)
# Performance
# List: Since lists are mutable, they tend to be slightly slower for certain operations (especially modification operations).
# Tuple: Being immutable, tuples are faster for iteration and have a smaller memory footprint.
# Use Cases
# List: When we need a collection that can change over time (e.g., when adding, removing, or updating elements).
# Tuple: When we need a fixed collection of elements that should not change (e.g., representing coordinates, returning multiple values from a function).
# Example:
# List: Modifiable
shopping_list = ["apple", "banana", "carrot"]
shopping_list.append("date")  # Adds "date" to the list
print(shopping_list)

# Tuple: Fixed data
coordinates = (10.5, 20.3)
coordinates [0] = 15 # Error! Tuples cannot be modified
# Methods
# List: A list has many built-in methods like append (), remove (), pop (), etc.
# Tuple: A tuple has limited methods, mainly count () and index ().
# Example:
my_list = [1, 2, 3]
my_list.append (4)  # Works, list can be modified

my_tuple = (1, 2, 3)
my_tuple.append (4)  # Error! Tuples don't have append ()
# Length
# List: A list can grow or shrink dynamically as you add or remove items.
# Tuple: It has fixed size; the number of elements is defined when it's created.
# Nestability
# Both lists and tuples can be nested within each other (i.e., you can have a list of tuples or a tuple of lists).
# Example:
nested_list = [[1, 2], [3, 4]]
nested_tuple = ((1, 2), (3, 4))


# Question 5: Describe the key features of sets and provide examples of their use?
# Answer 5: A set in Python is an unordered collection of unique elements. It is useful when we want to store a collection of distinct items and perform operations like union, intersection, and difference.
Below are the key features of sets:

# Unordered:
# The order of elements is not maintained in sets. The order of elements may differ from the insertion order, when we print or iterate through a set.

my_set = {3, 1, 2}
print(my_set)

# Unique Elements:
# A set automatically removes any duplicate elements. All the elements are unique in a set.

my_set = {1, 2, 2, 3}
print(my_set)

# Mutable
# The elements in a set although must be immutable (e.g., numbers, strings, tuples), the set itself is mutable. After creation we can add or remove elements.

# Add an element:
   my_set = {1, 2}
   my_set.add(3)
   print(my_set)

# Remove an element:
   my_set.remove(2)
   print(my_set)

# No Indexing
# Sets are unordered, so they do not support indexing or slicing like lists or tuples.

my_set [0] # Error! Sets cannot be indexed

# Set Operations:
# Python sets support mathematical set operations like union, intersection, difference, and symmetric difference.

# Union (`|`): Combines all elements from both sets, excluding duplicates.

   set1 = {1, 2, 3}
   set2 = {3, 4, 5}
   print (set1 | set2)

# Intersection (`&`): Returns only the elements common to both sets.
   print (set1 & set2)

# Difference (`-`): Returns elements in the first set but not in the second.
   print (set1 - set2)

# Symmetric Difference (`^`): Returns elements in either set, but not in both.
   print (set1 ^ set2)

# Iterating Over a Set:
# You can iterate over the elements of a set using a loop.

for item in my_set:
    print(item)

# Set Comprehension:
# Like list comprehensions, Python allows set comprehensions for creating sets in a concise way.

squared_set = {x**2 for x in range (5)}
print(squared_set)

# Example Use Cases:
#Removing Duplicates: Sets are commonly used to remove duplicates from a list.
   my_list = [1, 2, 2, 3, 4, 4]
   unique_set = set(my_list)
   print(unique_set)

# Checking Membership:
# Sets are highly optimized for checking whether an item is in the set.
   if 3 in my_set:
       print ("3 is in the set")

# Question 6: Discuss the use cases of tuples and sets in Python programming?
#Answer 6: Use Cases of Tuples:
# Fixed Data: When we want data that shouldn’t change, like days of the week or coordinates we use tuples.
#Example:
days = ("Mon", "Tue", "Wed")
# Multiple Return Values: Tuples are great for returning more than one value from a function.
# Example:
def get_coordinates ():
    return (10, 20)
# Dictionary Keys: Tuples can be used as keys in a dictionary since they’re immutable.
#Example:
my_dict = {(1, 2): "point"}
# Efficient Memory: Tuples are lighter than lists, so they’re useful for large, fixed datasets.
# Use Cases of Sets:
# Unique Elements: Sets automatically remove duplicates, making them perfect for keeping unique items.
# Example:
my_set = {1, 2, 2, 3}
# Set Operations: Great for union, intersection, and difference — like finding common or distinct items.
# Example:
set1 = {1, 2, 3}
set2 = {3, 4, 5}
common = set1 & set2
# Fast Membership Testing: Sets are much quicker than lists when you need to check if something exists.
# Example:
if 3 in my_set:
    print("Found!")
# Removing Duplicates: Easily remove duplicates from lists or large datasets.
# Example:
unique_names = set (["Alice", "Bob", "Alice"])

# Question 7: Describe how to add, modify, and delete items in a dictionary with examples?
# Answer 7: Dictionary: It is a data structure that stores data and is an unordered collection of key-value pairs. Each key in a dictionary is unique and is used to access the corresponding value. Dictionaries are mutable, meaning we can add, modify, and delete key-value pairs after the dictionary is created.

# Adding Items to a Dictionary
# If we want to add a new key-value pair, simply assign a value to a new key.
# Example:
my_dict = {"name": "Alice", "age": 25}
my_dict["city"] = "New York” # Adds new key-value pair
print(my_dict)
# Modifying Items in a Dictionary
# For modifying an existing item, assign a new value to an existing key.
# Example:
my_dict["age"] = 26 # Modifies the value of the 'age' key
print(my_dict)
# Deleting Items from a Dictionary
# We can delete a key-value pair using the del keyword or the pop() method.
# Using del:
del my_dict["city”] # Deletes the key 'city'
print(my_dict)
# Using pop (): This removes the key and returns its value.
removed_value = my_dict.pop("age”) # Removes 'age' and returns its value
print(my_dict)
print(removed_value)
# Below is an example for all the methods:
my_dict = {"name": "Alice", "age": 25}
my_dict["city"] = "New York” # Adding
my_dict["age"] = 26 # Modifying
del my_dict["city”] # Deleting


#Question 8: Discuss the importance of dictionary keys being immutable and provide examples?
#Answer 8: In Python, dictionary keys must be immutable because they need to remain constant throughout the lifecycle of the dictionary. Objects like strings, numbers, and tuples which are immutable ensure that keys maintain their hash values, which allows Python to efficiently look up, store, and retrieve key-value pairs.
# Reasons why Dictionary Keys must be immutable:
# Hashing Requirement:
# Dictionary keys are stored based on their hash values. It helps Python quickly find and retrieve the associated value.
# If keys were mutable (e.g., lists), their hash value could change after modification, making it impossible to reliably find or retrieve the value associated with the key.
# Efficient Lookup:
# Python uses a hash table for dictionaries, allowing constant time complexity (O (1)) for lookups. Immutable keys ensure that the lookup remains consistent and efficient, even as the dictionary grows.
# Key Stability:
# If keys were mutable and changed over time, it would break the dictionary’s internal structure, leading to errors or incorrect lookups.
# Examples of Immutable Keys:
# Strings:
my_dict = {"name": "Alice", "age": 25}
print(my_dict["name"])
# Strings, being immutable, are often used as dictionary keys.
# Numbers:
my_dict = {1: "One", 2: "Two"}
print(my_dict[1])
# Numbers are immutable, making them valid dictionary keys.
# Tuples (if they contain only immutable elements):
my_dict = {(1, 2): "Coordinates", (3, 4): "Point"}
print (my_dict[(1, 2)])
# Invalid Example (Mutable Key):
# Lists cannot be used as keys because they are mutable.
my_dict = {[1, 2]: "Invalid Key"}
```
